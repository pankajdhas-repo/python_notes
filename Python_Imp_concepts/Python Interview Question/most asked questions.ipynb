# üöÄ 11. Find Missing Number in Sequence

````python
#Find Missing Number in Sequence

def find_missing_number(num_list, n):
    """ logic to find missing number in a sequence is as follows 
     1. Calculate the expected sum of first n natural numbers using the formula n*(n+1)/2
     2. Calculate the actual sum of the numbers present in the list
     3. The missing number is the difference between the expected sum and the actual sum
    """
    expected_sum = n * (n + 1) // 2 # formula to calculate sum of first n natural numbers
    actual_sum = sum(num_list) 
    return expected_sum - actual_sum

find_missing_number([1, 2, 4, 5, 6], 6)
````

# üöÄ 12. Check Anagram

````python
# check anagram
def are_anagrams(str1, str2):
    return sorted(str1) == sorted(str2)

are_anagrams("listen", "silent")
````

````python
# with programming logic
def are_anagrams(str1, str2):
    """logic to check anagram is as follows
    1. If the lengths of the two strings are different, they cannot be anagrams.
    2. Create a dictionary to count the frequency of each character in the first string.
    3. Iterate through the second string, decrementing the character counts in the dictionary.
    4. If any character in the second string is not found in the dictionary or its count goes below zero, the strings are not anagrams.
    5. If all character counts return to zero, the strings are anagrams."""
    if len(str1) != len(str2):
        return False

    char_count = {}

    for char in str1:
        char_count[char] = char_count.get(char, 0) + 1

    for char in str2:
        if char not in char_count or char_count[char] == 0:
            return False
        char_count[char] -= 1

    return True

are_anagrams("listen", "silent")
````

# ‚úÖ **1. Singleton Pattern ‚Äî Only ONE instance allowed**

### **Meaning**

A class is Singleton when only **one object/instance** can ever be created from it.

### **Why needed?**

Some components must exist only once in the system:

* One database connection for whole app
* One logging manager
* One configuration reader
* One cache object

Creating multiple objects may cause memory problems or inconsistent states.

### **How it works?**

Python allows you to override object creation using `__new__()`.
We store the instance in a class variable.

### **Code (with explanation)**

```python
class Singleton:
    """
    Singleton pattern ensures only one instance exists.
    """
    _instance = None  # stores the single object

    def __new__(cls):
        if cls._instance is None:      # if no instance is created
            cls._instance = super().__new__(cls)
        return cls._instance           # return same instance every time
```

### **Explanation (step-by-step)**

1. Python calls `__new__()` *before* `__init__()`.
2. First time ‚Üí `_instance` is `None` ‚Üí new object created.
3. Second time ‚Üí `_instance` already has object ‚Üí returns same object.

So both `a` and `b` point to the same memory location.

Instead of doing:

### **Real-world example**

You select *Auto*, *Bike*, or *Car*.
You don‚Äôt create the vehicle object.

### **Flow**

3. User gets object without knowing creation details

## **S ‚Äî Single Responsibility Principle (SRP)**

**Technical meaning:**
A class should **do only one thing**. Don‚Äôt mix multiple jobs in the same class.

**Analogy for a school boy:**
Imagine a **teacher**. The teacher teaches **math** only.
If the teacher also cooks lunch, fixes the classroom, and drives the bus, it becomes messy.

**Python Example**

```python
# ‚ùå Bad ‚Äî doing too many things

    def __init__(self, name):
        self.name = name

    def attend_class(self):
        print(f"{self.name} is attending class")

    def submit_homework(self):
        print(f"{self.name} submits homework")

    def pay_fees(self):
        print(f"{self.name} pays fees")  # Too many responsibilities
```

```python
# ‚úÖ Good ‚Äî single responsibility

    def mark(self, student):

    def submit(self, student):
        print(f"{student} submitted homework")

    def pay(self, student):
```

## **L ‚Äî Liskov Substitution Principle (LSP)**

**Technical meaning:**
If **B is a child of A**, B should behave like A.
You should be able to replace A with B without breaking the code.

**Analogy:**
A **car** is a vehicle. If you replace the car with a **bike**, it should still move on the road like a vehicle.
If your ‚Äúbike‚Äù can‚Äôt move or breaks all rules of a vehicle ‚Üí it violates LSP.

**Python Example**

```python

    pass

    def fly(self):
        raise Exception("Ostrich can't fly")  # ‚ùå violates LSP
```

**Better design:**

```python

    def eat(self):
        print("Bird is eating")

    def fly(self):
        print("Flying bird can fly")

class Eagle(FlyingBird):
    pass

class Ostrich(Bird):
    pass  # no fly method, still works fine
```

**Key idea:** Child class should **behave like parent**.

* Call
* Camera
* GPS

````python
def cache_results(func):
    # dictionary to store cached results
    cache = {}

    def wrapper(num):
        # if num is already in cache, return cached value
        if num in cache:
            print("Returning from cache...")
            return cache[num]

        # otherwise, compute result and store in cache
        result = func(num)
        cache[num] = result
        print("Computed and cached result...")
        return result

    return wrapper

@cache_results
def square(n):
    return n * n

print(square(4))
print(square(5))
print(square(4))
print(square(6))
print(square(7))
print(square(5))
````

# ‚≠ê 1. What is an LRU Cache?

LRU = **Least Recently Used**

It means:

* We keep only a limited number of cached values
* When the cache becomes full, we remove the **least recently used** item

Example:
Cache size = 3
Sequence: `3, 2, 1, 4`

Before inserting 4, the oldest (least used) value is removed.

---

## **Interview-Ready Definition**

> ‚ÄúAn LRU Cache stores recently used results to avoid recomputation.
> It has a fixed capacity, and when full, it evicts the least recently used item.
> This improves performance while keeping memory usage controlled.‚Äù

---

# ‚≠ê 2. Using Python's Built-in LRU Cache

Just add a decorator:

```python
from functools import lru_cache

@lru_cache(maxsize=5)
def square(n):
    print("Computing...")
    return n * n

print(square(4))  # computed
print(square(4))  # cached
```

### Output

```
Computing...
16
16  # from cache (no computing)
```

‚úî Thread-safe
‚úî Automatically handles eviction
‚úî Very efficient

---

# ‚≠ê 3. Make Your Own LRU Cache (Custom Implementation)

This is a **real interview-level coding question**.

We can use:

* **OrderedDict** (preserves insertion order)
* Moves items when accessed

---

### ‚úÖ Custom LRU Cache Decorator

```python
from collections import OrderedDict
from functools import wraps

def lru_cache_custom(maxsize=3):
    def decorator(func):
        cache = OrderedDict()

        @wraps(func)
        def wrapper(*args):
            if args in cache:
                cache.move_to_end(args)  # mark as recently used
                print("Returning from cache:", args)
                return cache[args]

            result = func(*args)
            cache[args] = result
            print("Computed:", args)

            if len(cache) > maxsize:
                cache.popitem(last=False)  # remove least recently used
                print("Evicted least recently used")

            return result

        return wrapper
    return decorator
```

---

### Use it:

```python
@lru_cache_custom(maxsize=3)
def square(n):
    return n * n

square(2)
square(3)
square(4)
square(2)  # cached
square(5)  # triggers eviction
```

# ‚≠ê 4. When NOT to Use Caching

Caching may cause problems when:

‚ùå Function returns **different results each time** (e.g., random numbers, time-based output)
‚ùå Function performs **write operations** (database writes, file writes)
‚ùå Input space is VERY large (cache will grow huge)
‚ùå Used in memory-constrained environments
‚ùå Sensitive data storage (cache may reveal user data)

---

# ‚≠ê 5. Time Complexity Benefits

| Operation     | Without Cache   | With Cache             |
| ------------- | --------------- | ---------------------- |
| First call    | O(n) or heavy   | same heavy computation |
| Repeated call | O(n) every time | **O(1)** lookup        |
| Memory usage  | Low             | Higher (cache stored)  |

````python
def flatten(lst):
    """ flattens a nested list into a single list """
    result = []
    for item in lst:
        if isinstance(item, list):
            result.extend(flatten(item))
        else:
            result.append(item)
    return result

list_to_flatten = [1, [2, [3, 4], 5], 6]
flattened_list = flatten(list_to_flatten)
print(flattened_list)
````

````python
def flatten_generator(lst):
    for item in lst:
        if isinstance(item, list):
            yield from flatten_generator(item)
        else:
            yield item

list_to_flatten = [1, [2, [3, 4], 5], 6]
flattened_list = list(flatten_generator(list_to_flatten))
print(flattened_list)
````

````python
def count_without_extra(arr):
    arr.sort()
    count = 1
    result = []

    for i in range(1, len(arr)):
        if arr[i] == arr[i-1]:
            count += 1
        else:
            result.append((arr[i-1], count))
            count = 1

    result.append((arr[-1], count))  # last element
    return result


count_without_extra([1, 2, 2, 3, 4, 4, 5])
````

````python

import re

def capitalize_proper(text: str) -> str:
    """Capitalize each word like a proper phrase."""
    return " ".join(word.capitalize() for word in text.split())

def normalize_phrase(s: str) -> str:
    # Regex to match "navigate to <place>" in any casing/spelling
    pattern = re.compile(r"(navigate\s+to)\s+(.*)", re.IGNORECASE)

    match = pattern.match(s.strip())
    if not match:
        return capitalize_proper(s)  # fallback

    command = match.group(1)        # "navigate to"
    destination = match.group(2)    # "<place>"

    # Normalize both parts
    normalized_command = capitalize_proper(command)
    normalized_destination = capitalize_proper(destination)

    return f"{normalized_command} {normalized_destination}"


print(normalize_phrase("Navigate to starbucks"))
print(normalize_phrase("navigate to sTarbUcks"))
print(normalize_phrase("NAVIGATE to STARBUCKS"))

````

````python
class LRUCache:
    def __init__(self, max_size=3):
        self.cache = {}
        self.order = []
        self.max_size = max_size

    def get(self, key):
        if key in self.cache:
            # move key to end (most recently used)
            self.order.remove(key)
            self.order.append(key)
            return self.cache[key]
        return None

    def put(self, key, value):
        if key in self.cache:
            # update existing key
            self.order.remove(key)
        elif len(self.cache) == self.max_size:
            # evict LRU
            lru = self.order.pop(0)
            del self.cache[lru]

        self.cache[key] = value
        self.order.append(key)

# Example usage
lru = LRUCache(3)
lru.put("a", 1)
lru.put("b", 2)
lru.put("c", 3)
lru.get("a")      # a becomes MRU
lru.put("d", 4)   # evicts b
print(lru.cache)  # {'a':1, 'c':3, 'd':4}
````

````python
#context manager 

class FileHandler:
    """Context manager for file handling as follows
    1. The __init__ method initializes the FileHandler with the filename.
    2. The __enter__ method opens the file and returns the file object.
    3. The __exit__ method closes the file, ensuring proper resource management.
    """
    def __init__(self, filename):
        self.filename = filename
    
    def __enter__(self):
        self.file = open(self.filename, "r")
        return self.file
    
    def __exit__(self, exc_type, exc_val, exc_tb): 
        # exc_type, exc_val, exc_tb -- to handle exceptions if any
        # exc_type: exception type
        # exc_val: exception value
        # exc_tb: traceback object which shows where exception occurred
        self.file.close()

with FileHandler("data.txt") as f:
    print(f.read())
````

# üß† **Memory Management in Python ‚Äî Technical Explanation**

Python‚Äôs memory management is a combination of **automatic memory handling**, **reference counting**, **garbage collection**, and **allocator layers** optimized for performance.

It is mainly governed by:

1. **Reference Counting** (primary mechanism)
2. **Garbage Collection (GC)** for cyclic references
3. **Memory Allocators (pymalloc + OS allocators)**
4. **Object Interning**
5. **Memory Pools / Arenas**

Let's break each down.

---

# ‚≠ê **1. Reference Counting (Primary Memory Management System)**

Every Python object has an internal counter that tracks how many references point to it.

* When an object is created ‚Üí refcount = 1
* When a new variable references it ‚Üí refcount += 1
* When a reference is deleted ‚Üí refcount -= 1
* When refcount reaches **0** ‚Üí memory is freed **immediately**

### Example:

```python
import sys

a = []
sys.getrefcount(a)  # Shows reference count
```

### Why python uses reference counting?

* Simple & fast
* Immediate deallocation (vs waiting for GC)
* Predictable memory release

### Weakness:

Cannot handle **cyclic references** (A ‚Üí B ‚Üí A)

This is why garbage collector is needed.

---

# ‚≠ê **2. Garbage Collection (GC) ‚Äî Cyclic Garbage Collector**

Python uses a **generational garbage collector** to handle objects that reference each other.

### **Three Generations**

Python divides objects into:

* Gen 0 (youngest)
* Gen 1
* Gen 2 (oldest)

Objects surviving multiple collections are promoted to higher generations.

### Why use generations?

Most objects die young ‚Üí generational GC improves performance.

### How GC detects garbage?

1. GC scans containers (lists, dicts, classes)
2. Finds objects that are **only referenced within a cycle**
3. If the cycle has no external references ‚Üí objects are collected

### Trigger conditions:

* Object allocations exceed thresholds
* Manual call: `gc.collect()`

---

# ‚≠ê **3. Python Memory Allocators (pymalloc)**

Python uses several layers:

## **a) pymalloc**

Optimized allocator for **small objects (<512 bytes)**:

* Very fast
* Reduces fragmentation
* Uses memory pools & blocks

## **b) Arena-based allocation**

* Memory is grouped into **arenas** (256 KB)
* Arenas ‚Üí pools ‚Üí blocks
* When an arena is empty ‚Üí returned to OS

## **c) For large objects**

Python delegates allocation to the **C-level malloc()**, controlled by the OS.

This improves:
‚úî speed
‚úî memory reuse
‚úî fragmentation control

---

# ‚≠ê **4. Object Interning (Optimization Technique)**

Python **reuses** common immutable objects to save memory.

### What gets interned?

#### Automatically:

* Small integers (usually -5 to 256)
* Short strings
* Identifiers (variable names)

#### Benefits:

* Faster object creation
* Saves memory
* Fast comparison using pointer equality

Example:

```python
a = 100
b = 100
print(a is b)   # True because of interning
```

---

# ‚≠ê **5. Stack vs Heap Memory in Python**

### Stack:

* Stores local variable references
* Function call frames
* Very fast but small

### Heap:

* Stores actual Python objects (lists, dicts, user objects)
* Managed by Python memory manager
* Slower but large

Python variables are **references stored on the stack**, while **objects live on the heap**.

---

# ‚≠ê **6. Memory Leaks in Python & How They Occur**

Python is protected against most leaks due to automatic cleanup, BUT leaks can still occur:

### Common causes:

1. **Global variables never cleared**
2. **Unbounded caches**
3. **Lingering references in containers**
4. **Cyclic references with `__del__()` methods**
5. **Improper use of C extensions**

---

# ‚≠ê **7. Tools to Monitor Memory Usage**

### Built-in:

```python

sys.getsizeof(obj)
```

### External tools:

* **tracemalloc**
* **memory_profiler**
* **objgraph**
* **PySpy**

---

# ‚≠ê **8. Memory Optimization Techniques**

* Use generators instead of lists
* Use `__slots__` to reduce per-object overhead
* Avoid unnecessary object creation
* Use built-in types (they‚Äôre optimized in C)
* Use weak references (`weakref` module)

‚úî pymalloc memory allocator (pools, arenas)

‚úî Stack for references, heap for objects

Absolutely! Let‚Äôs go **step by step** on how Python achieves **multithreading** using `ThreadPoolExecutor` (from `concurrent.futures`) and how it works under the hood.

---

# üß† **1. What is ThreadPoolExecutor?**

`ThreadPoolExecutor` is a **high-level API** for running tasks concurrently using a **pool of threads**.

* You **submit tasks** (functions) to the pool.
* The executor **manages threads**, schedules tasks, and executes them.
* It abstracts away manual `threading.Thread` creation.

**Modules used:**

```python
from concurrent.futures import ThreadPoolExecutor, as_completed
```

---

# ‚≠ê **2. Example: Using ThreadPoolExecutor**

Suppose we have an I/O-bound task like downloading a file:

```python
import time
from concurrent.futures import ThreadPoolExecutor

def download_file(file_id):
    print(f"Downloading file {file_id}")
    time.sleep(2)  # Simulate I/O delay
    return f"File {file_id} downloaded"

# Create a ThreadPoolExecutor with 5 threads
with ThreadPoolExecutor(max_workers=5) as executor:
    # Submit 10 download tasks
    futures = [executor.submit(download_file, i) for i in range(10)]
    
    # Wait for all tasks to complete and print results
    for future in futures:
        print(future.result())
```

### **How it works:**

1. `max_workers=5` ‚Üí 5 threads are created in a pool.
2. `executor.submit()` ‚Üí schedules the function for execution.
3. The pool assigns tasks to threads **concurrently**.
4. Threads **release the GIL** during I/O (`time.sleep`, file/network operations).
5. Results are collected via `Future` objects.

---

# ‚≠ê **3. Using `map()` with ThreadPoolExecutor**

A simpler version using `executor.map`:

```python
with ThreadPoolExecutor(max_workers=5) as executor:
    results = executor.map(download_file, range(10))

for r in results:
    print(r)
```

* `map()` automatically waits for all threads to finish.
* Keeps order of inputs.

---

# ‚≠ê **4. Achieving Multithreading**

* Threads in `ThreadPoolExecutor` share the **same process memory space**.
* Python **creates multiple threads** internally.
* Each thread can run concurrently if the task is I/O-bound because Python **releases the GIL** during I/O operations (network calls, file I/O, sleep, etc.).

**Important:** CPU-bound tasks won‚Äôt speed up because GIL allows only **one thread to execute Python bytecode at a time**.

---

# ‚≠ê **5. Example: Mixing CPU and I/O Tasks**

```python
from concurrent.futures import ThreadPoolExecutor
import math, time

def io_task(n):
    time.sleep(1)
    return f"IO {n} done"

def cpu_task(n):
    for _ in range(1000000):
        math.sqrt(n)
    return f"CPU {n} done"

with ThreadPoolExecutor(max_workers=3) as executor:
    futures = [executor.submit(io_task, i) for i in range(3)]
    futures += [executor.submit(cpu_task, i) for i in range(3)]

for f in futures:
    print(f.result())
```

* `io_task` will benefit from threading ‚Üí runs concurrently
* `cpu_task` will **not** benefit ‚Üí GIL blocks true parallel execution

---

# ‚≠ê **6. Key Points**

1. `ThreadPoolExecutor` manages **thread lifecycle** automatically.
2. You **submit tasks**, which return `Future` objects.
3. Python releases the GIL for **I/O-bound operations**, allowing **concurrency**.
4. For CPU-bound operations ‚Üí use `ProcessPoolExecutor` (multiprocessing version).
5. Maximum threads are controlled by `max_workers`.

---

# üéØ **Interview-ready explanation**

> In Python, we can achieve multithreading using `concurrent.futures.ThreadPoolExecutor` by submitting tasks to a pool of threads. Each thread executes a function concurrently. Python threads are effective for I/O-bound operations because the GIL is released during I/O, allowing multiple threads to run concurrently. For CPU-bound tasks, `ThreadPoolExecutor` does not achieve true parallelism; `ProcessPoolExecutor` should be used instead.
