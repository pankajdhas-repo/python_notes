ğŸ”¹ What is a Lambda Function?

A lambda function in Python is a small anonymous function â€” meaning it doesnâ€™t have a name (unlike functions you create with def).
It can take any number of arguments, but it can only contain one expression.

ğŸ”¹ Syntax:
lambda arguments: expression

````python
# Regular function
def add(a, b):
    return a + b

# Lambda equivalent
add_lambda = lambda a, b: a + b

print(add(3, 5))        
print(add_lambda(3, 5))
````

````python
check_even_odd = lambda num: "Even" if num % 2 == 0 else "Odd"

print(check_even_odd(4))  
print(check_even_odd(7)) 
````

````python
largest = lambda a, b: a if a > b else b

print(largest(10, 25))
````

````python
grade = lambda marks: "Pass" if marks >= 40 else "Fail"

print(grade(85)) 
print(grade(30))
````

````python
numbers = [1, 2, 3, 4, 5]
doubled = list(map(lambda x: x * 2, numbers))
print(doubled)
````

````python
even_nums = list(filter(lambda x: x % 2 == 0, numbers))
print(even_nums)
````

````python
odd_nums = list(filter(lambda x: x % 2 != 0, numbers))
print(odd_nums)
````

````python
grade = lambda m: "Excellent" if m >= 80 else ("Good" if m >= 60 else ("Pass" if m >= 40 else "Fail"))

marks = int(input("Enter your marks: "))
print("Result:", grade(marks))

````

````python
check_sign = lambda x: "Positive" if x > 0 else ("Negative" if x < 0 else "Zero")

num = int(input("Enter a number: "))
print("The number is:", check_sign(num))
````

````python
leap = lambda y: "Leap Year" if (y % 400 == 0 or (y % 4 == 0 and y % 100 != 0)) else "Not Leap Year"

year = int(input("Enter a year: "))
print(leap(year))
````

````python
smallest = lambda a, b, c: a if a < b and a < c else (b if b < c else c)

a = 54
b = 45
c = 33
print("Smallest number is:", smallest(a, b, c))
````

````python
from functools import reduce

nums = [1, 2, 3, 4, 5]
total = reduce(lambda a, b: a + b, nums)
print("Sum:", total)
````

````python
nums = [1, 2, 3, 4, 5, 6]

# Step 1: filter even numbers
# Step 2: map to squares
# Step 3: reduce to sum
result = reduce(lambda a, b: a + b, map(lambda x: x ** 2, filter(lambda x: x % 2 == 0, nums)))
print("Sum of even squares:", result)
````

## ğŸ§¾ **Comparison: `map()` vs `filter()` vs `reduce()`**

| Feature                   | **map()**                                                                       | **filter()**                                                     | **reduce()**                                       |
| ------------------------- | ------------------------------------------------------------------------------- | ---------------------------------------------------------------- | -------------------------------------------------- |
| **Purpose**               | Transforms each element in an iterable                                          | Filters elements based on a condition                            | Reduces all elements to a single result            |
| **Returns**               | A new iterable (same length as input)                                           | A filtered iterable (smaller or equal length)                    | A single value (number, string, etc.)              |
| **Function type**         | Function returns **transformed value**                                          | Function returns **True/False**                                  | Function combines **two values** at a time         |
| **Import needed?**        | âŒ No                                                                            | âŒ No                                                             | âœ… Yes â€” from `functools`                           |
| **Typical Use Cases**     | - Change format or data type<br>- Square or double numbers<br>- Convert strings | - Keep only certain elements<br>- Filter evens, positives, words | - Find sum, product, min, max, or combine all data |
| **Example with `lambda`** | `map(lambda x: x * 2, [1,2,3])` â†’ `[2,4,6]`                                     | `filter(lambda x: x % 2 == 0, [1,2,3,4])` â†’ `[2,4]`              | `reduce(lambda a,b: a + b, [1,2,3,4])` â†’ `10`      |
| **Output Type**           | `map` object â†’ convert to `list()`                                              | `filter` object â†’ convert to `list()`                            | Single value (int, float, etc.)                    |
| **Can be combined with**  | `filter()`, `reduce()`                                                          | `map()`, `reduce()`                                              | `map()`, `filter()`                                |
| **Best For**              | Element-wise transformations                                                    | Conditional selections                                           | Aggregations and summaries                         |

## âš™ï¸ **Example: Using All Three Together**

Letâ€™s combine all three in one program ğŸ‘‡

```python
from functools import reduce

nums = [1, 2, 3, 4, 5, 6]

# Step 1: Filter even numbers
evens = filter(lambda x: x % 2 == 0, nums)

# Step 2: Square them
squares = map(lambda x: x ** 2, evens)

# Step 3: Add them all together

print("Sum of even squares:", result)
```
