```python
# Reverse a string 
def reverse_string(s):
    reversed_str = ""
    for char in s:
        reversed_str = char + reversed_str
    return reversed_str

rev_str = reverse_string('ganesh')
print(f'reverse string with loop --> ',rev_str)

def reverse_string_builtin(s):
    return ''.join(reversed(s))

rev_inbuilt_str = reverse_string_builtin('bhalekar')
print(f'reverse string with inuilt func --> {rev_inbuilt_str}')
```

```python
# Find the second largest number in a list

def second_largest(nums):
    largest = second = float('-inf')
    for num in nums:
        if num > largest:
            second = largest
            largest = num
        elif largest > num > second:
            second = num
    return second if second != float('-inf') else None

num_list = [76,32,17,68,9,43]
second_largest_ele = second_largest(num_list)
print(f'from the input list {num_list}--> the second largest number is {second_largest_ele}')
```

```python
def second_largest_with_built_functions(nums):
    unique_nums = list(set(nums))
    if len(unique_nums) < 2:
        return None
    unique_nums.sort()
    return unique_nums[-2]

num_list = [76,32,17,68,9,43]
second_largest_ele = second_largest_with_built_functions(num_list)
print(f'from the input list {num_list}--> the second largest number is {second_largest_ele}')
```

```python
#Count the frequency of each character in a string

def char_frequency(s):
    freq = {}
    for char in s:
        if char in freq:
            freq[char] += 1
        else:
            freq[char] = 1
    return freq

s = 'this is test'
char_frequency(s)
```

```python
from collections import Counter

def char_frequency_builtin(s):
    return Counter(s)

s = 'this is test'
char_frequency_builtin(s)
```

```python
#Check if a number is prime

def is_prime(n):
    if n <= 1:
        return False
    for i in range(2, n):
        if n % i == 0:
            return False
    return True

is_prime(21)
```

```python
import math

def is_prime_optimized(n):
    if n <= 1:
        return False
    for i in range(2, int(math.sqrt(n)) + 1):
        if n % i == 0:
            return False
    return True

is_prime_optimized(21)
```

```python
#Merge two sorted lists into one sorted list
def merge_sorted_lists(a, b):
    result = []
    i = j = 0
    while i < len(a) and j < len(b):
        if a[i] < b[j]:
            result.append(a[i])
            i += 1
        else:
            result.append(b[j])
            j += 1
    # Append remaining elements
    result.extend(a[i:])
    result.extend(b[j:])
    return result

list1 = [1,2,3,4,5]
list2 = [6,7,8,9,0]
merge_sorted_lists(list1,list2)
```

```python
def merge_sorted_lists_builtin(a, b):
    return sorted(a + b)

list1 = [1,2,3,4,5]
list2 = [6,7,8,9,0]
merge_sorted_lists_builtin(list1,list2)
```

```python
#Remove duplicates from a list while preserving order

def remove_duplicates(lst):
    """ Used two lists to preserve order one is for checking and another for result """
    seen = []
    result = []
    for item in lst:
        if item not in seen:
            seen.append(item)
            result.append(item)
    return result

input_ls = [0, 1, 2, 3, 4, 1, 6, 2, 8, 4]
remove_duplicates(input_ls)
```

```python
#Check if a string is a palindrome

def is_palindrome(s):
    left, right = 0, len(s) - 1
    while left < right:
        if s[left] != s[right]:
            return False
        left += 1
        right -= 1
    return True
s = 'bob'
is_palindrome(s)
```

```python
def is_palindrome_builtin(s):
    return s == s[::-1]
s = 'bob'
is_palindrome_builtin(s)
```

```python
# palindrome with using for loop
def is_palindrome_for(s):
    n = len(s)
    for i in range(n // 2):
        if s[i] != s[n - i - 1]: #compare characters from start and end
            return False
    return True
s = 'madam'
is_palindrome_for(s)
```

```python
#Find the factorial of a number using recursion
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

factorial(5)
```

```python
#Generate Fibonacci series up to n terms
def fibonacci_series(n):
    """ generate fibonacci series up to n terms 
    logic: 
    define two variables a and b to hold the last two numbers in the series
    iterate n times, appending a to the series and updating a and b
    returns a list containing the Fibonacci series up to n terms
    """
    fib = []
    a, b = 0, 1
    for _ in range(n):
        fib.append(a)
        a, b = b, a + b
    return fib

fibonacci_series(10)
```

```python
def fibonacci_recursive(n):
    if n <= 1:
        return n
    return fibonacci_recursive(n-1) + fibonacci_recursive(n-2)

def fibonacci_list(n):
    return [fibonacci_recursive(i) for i in range(n)]

fibonacci_list(10)
```

```python
#Find common elements between two lists
def common_elements(a, b):
    return list(set(a) & set(b))

list1 = [1,2,3,4,5]
list2 = [6,7,2,3,0]
common_elements(list1,list2)
```

```python
##Find common elements between two lists
def common_elements_manual(a, b):
    result = []
    for item in a:
        # check if item is in both lists and not already in result then append to result
        if item in b and item not in result:
            result.append(item)
    return result

list1 = [1,2,3,4,5]
list2 = [6,7,2,3,0]
common_elements_manual(list1,list2)
```

```python
def length_of_longest_substring(s):
    # Set to store unique characters in the current window (substring)
    char_set = set()

    # Left pointer of the sliding window
    left = 0

    # Variable to keep track of the maximum length found
    max_length = 0

    # Iterate through the string with the right pointer
    for right in range(len(s)):
        # If the character at 'right' is already in the set (duplicate)
        # then shrink the window from the left until it's removed
        while s[right] in char_set:
            # Remove the character at the 'left' pointer from the set
            char_set.remove(s[left])
            # Move the left pointer to the right
            left += 1

        # Add the current character to the set (no duplicate now)
        char_set.add(s[right])

        # Update the max_length with the size of the current window
        max_length = max(max_length, right - left + 1)

    # Return the length of the longest substring without duplicates
    return max_length


print(length_of_longest_substring("abcabcbb"))
```

```python
#Python code to implement a caching mechanism using a dictionary with max size = 3 (LRU)

from collections import OrderedDict

class LRUCache:
    def __init__(self, max_size=3):
        self.cache = OrderedDict()
        self.max_size = max_size

    def get(self, key):
        if key in self.cache:
            # Move key to end to show it was recently used
            self.cache.move_to_end(key)
            return self.cache[key]
        return None  # Or raise KeyError / return default

    def put(self, key, value):
        if key in self.cache:
            # Update value and mark as recently used
            self.cache.move_to_end(key)
        self.cache[key] = value

        if len(self.cache) > self.max_size:
            # Pop the first item (least recently used)
            evicted_key, evicted_value = self.cache.popitem(last=False)
            print(f"Evicted: ({evicted_key}, {evicted_value})")

    def __repr__(self):
        return f"{self.cache}"

# âœ… Example usage
cache = LRUCache(max_size=3)

cache.put('a', 1)
cache.put('b', 2)
cache.put('c', 3)
print(cache)  # a, b, c

cache.get('a')  # Access 'a' to make it recently used
cache.put('d', 4)  # This should evict 'b'
print(cache)  # c, a, d
```
